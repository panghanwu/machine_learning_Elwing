```
import warnings
# turn off FutureWarning
warnings.filterwarnings("ignore", category=FutureWarning)
```

```
from gensim.models.fasttext import FastText
# load
model = FastText.load_fasttext_format('cc.zh.300.bin')
```

```
len(model.wv.vocab)
```

```
model.wv.vocab
```

```
# Colab forms
text = 'value' #@param {type:"string"}
```

```
word = "hi"#@param {type:"string"}
print(word, "的詞向量:")
model.wv[word]
```

```
text = '\u5B85\u7537' #@param {type:"string"}
text
```

```
#@title 找出相似詞
text = '\u5B85\u7537' #@param {type:"string"}
model.wv.most_similar(text)
```

```
#@title 比較兩個特定詞的相似度

text1 = '台灣' #@param {type:"string"}
text2 = '台北' #@param {type:"string"}
model.wv.similarity(text1, text2)
```

```
#@title 找到不是同一掛的詞

text1 = '\u798F\u7279' #@param {type:"string"}
text2 = '\u8CD3\u58EB' #@param {type:"string"}
text3 = '\u88D5\u9686' #@param {type:"string"}
text4 = '\u4E09\u6D0B' #@param {type:"string"}

notmatch = model.wv.doesnt_match([text1, text2, text3, text4])
print("不是同一掛的:", notmatch)
```

```
#@title 詞相對關係

# text1 + text2 - text3 = ?
# text1 - text3 = ? - text2
text1 = '\u99AC\u82F1\u4E5D' #@param {type:"string"}
text2 = '\u67EF\u6587\u54F2' #@param {type:"string"}
text3 = '\u8521\u82F1\u6587' #@param {type:"string"}
question = model.wv.most_similar(positive=[text1, text2], negative=[text3])
print(text1, "-", text3, "=", "?", "-", text2)
question
```

有解決Out-Of-Vocabulary (OOV)問題

```
text = "是在哈囉"
print(text, "有在字典裡?", text in model.wv.vocab)
model.wv.similarity(text, "白癡")
```
