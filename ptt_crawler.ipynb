# Auto Crawler

```
start_page = 1607
pages = 10
index_link = 'https://www.ptt.cc/bbs/Soft_Job/index{page}.html'
articles = []

pgs = 0
# contents
for i in range(pages):
  response = requests.get(index_link.format(page=str(1607-i)),
               cookies={'over18':'1'})
  html = BeautifulSoup(response.text, 'lxml')
  article_list = html.find_all('div', class_='title')
  
  for art in article_list:
    try: 
      ann = '公告' in art.find('a').text
      if not ann:
        link = art.find('a')['href']
        content = requests.get('https://www.ptt.cc'+link,
                    cookies={'over18':'1'})
        content = BeautifulSoup(content.text, 'lxml')
        content = content.find('body').text
        articles.append(content)
    except:
      pass
  pgs += 1
  print('\r' + '[Crawling]:[%s%s]%.2f%%' % (
      '█' * int(pgs*30/pages), '  ' * (30-int(pgs*30/pages)),
      float(pgs/pages*100)), end='')

print('\nDone, total articles are', str(len(articles)) + '.')
```

```
import csv

with open('ptt_softjob.csv', 'w', newline='') as f:
     wr = csv.writer(f, quoting=csv.QUOTE_ALL)
     wr.writerow(articles)

articles[:10]
```
